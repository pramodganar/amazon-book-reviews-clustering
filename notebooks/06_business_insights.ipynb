```python
# ============================================================
# AMAZON BOOK REVIEWS CLUSTERING PROJECT
# NOTEBOOK: 06_business_insights.ipynb
# ============================================================

# ============================================================
# STEP 1 — IMPORT LIBRARIES
# ============================================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from collections import Counter

from wordcloud import WordCloud

import warnings
warnings.filterwarnings("ignore")

print("Libraries Imported Successfully")


# ============================================================
# STEP 2 — LOAD CLUSTERED DATASET
# ============================================================

DATA_PATH = (
    "../data/processed/"
    "clustered_reviews.csv"
)

df = pd.read_csv(DATA_PATH)

print("\nClustered Dataset Loaded Successfully")

print("\nDataset Shape:")
print(df.shape)


# ============================================================
# STEP 3 — CHECK DATASET
# ============================================================

print("\nDataset Columns:")

print(df.columns.tolist())

display(df.head())


# ============================================================
# STEP 4 — SELECT BEST CLUSTER MODEL
# ============================================================

# Using KMeans as primary clustering model

cluster_column = "kmeans_cluster"

print(f"\nUsing Cluster Column: {cluster_column}")


# ============================================================
# STEP 5 — CHECK CLUSTER DISTRIBUTION
# ============================================================

cluster_distribution = (
    df[cluster_column]
    .value_counts()
    .sort_index()
)

print("\nCluster Distribution:")

print(cluster_distribution)


# ============================================================
# STEP 6 — VISUALIZE CLUSTER DISTRIBUTION
# ============================================================

plt.figure(figsize=(10,6))

sns.countplot(
    x=cluster_column,
    data=df
)

plt.title("Customer Review Cluster Distribution")

plt.xlabel("Cluster")

plt.ylabel("Number of Reviews")

plt.show()


# ============================================================
# STEP 7 — TOP KEYWORDS PER CLUSTER
# ============================================================

print("\n======================================")
print("TOP KEYWORDS PER CLUSTER")
print("======================================")

for cluster in sorted(
    df[cluster_column].unique()
):

    cluster_text = " ".join(

        df[
            df[cluster_column] == cluster
        ]["cleaned_text"]
    )

    words = cluster_text.split()

    word_freq = Counter(words)

    top_words = word_freq.most_common(15)

    print(f"\nCluster {cluster}")

    print("-" * 40)

    print(top_words)


# ============================================================
# STEP 8 — GENERATE WORD CLOUDS
# ============================================================

print("\nGenerating Word Clouds...")

clusters = sorted(
    df[cluster_column].unique()
)

for cluster in clusters:

    cluster_text = " ".join(

        df[
            df[cluster_column] == cluster
        ]["cleaned_text"]
    )

    wordcloud = WordCloud(

        width=800,

        height=400,

        background_color="white"
    ).generate(cluster_text)

    plt.figure(figsize=(12,6))

    plt.imshow(wordcloud)

    plt.axis("off")

    plt.title(f"Word Cloud — Cluster {cluster}")

    plt.show()


# ============================================================
# STEP 9 — REVIEW LENGTH ANALYSIS
# ============================================================

plt.figure(figsize=(12,6))

sns.boxplot(

    x=cluster_column,

    y="review_length",

    data=df
)

plt.title("Review Length by Cluster")

plt.xlabel("Cluster")

plt.ylabel("Review Length")

plt.show()


# ============================================================
# STEP 10 — WORD COUNT ANALYSIS
# ============================================================

plt.figure(figsize=(12,6))

sns.boxplot(

    x=cluster_column,

    y="word_count",

    data=df
)

plt.title("Word Count by Cluster")

plt.xlabel("Cluster")

plt.ylabel("Word Count")

plt.show()


# ============================================================
# STEP 11 — TOP REVIEWERS PER CLUSTER
# ============================================================

print("\n======================================")
print("TOP REVIEWERS PER CLUSTER")
print("======================================")

for cluster in clusters:

    print(f"\nCluster {cluster}")

    print("-" * 40)

    top_reviewers = (

        df[
            df[cluster_column] == cluster
        ]["profilename"]

        .value_counts()

        .head(5)
    )

    print(top_reviewers)


# ============================================================
# STEP 12 — SAMPLE REVIEWS FROM EACH CLUSTER
# ============================================================

print("\n======================================")
print("SAMPLE REVIEWS PER CLUSTER")
print("======================================")

for cluster in clusters:

    print(f"\n\nCLUSTER {cluster}")

    print("=" * 60)

    sample_reviews = (

        df[
            df[cluster_column] == cluster
        ]["review_text"]

        .sample(
            3,
            random_state=42
        )
    )

    for idx, review in enumerate(sample_reviews):

        print(f"\nReview {idx+1}:")

        print(review[:500])

        print("\n" + "-"*50)


# ============================================================
# STEP 13 — CLUSTER INSIGHT GENERATION
# ============================================================

cluster_insights = {

    0: "Detailed Analytical Reviewers",

    1: "Emotional Readers",

    2: "Short Opinion Reviewers",

    3: "Highly Expressive Customers",

    4: "Book Enthusiasts"
}

print("\n======================================")
print("CLUSTER INSIGHTS")
print("======================================")

for cluster, insight in cluster_insights.items():

    print(f"\nCluster {cluster}: {insight}")


# ============================================================
# STEP 14 — CREATE CLUSTER INSIGHT COLUMN
# ============================================================

df["cluster_insight"] = (

    df[cluster_column]
    .map(cluster_insights)
)

print("\nCluster Insight Column Created")


# ============================================================
# STEP 15 — CLUSTER INSIGHT DISTRIBUTION
# ============================================================

plt.figure(figsize=(12,6))

sns.countplot(

    y="cluster_insight",

    data=df,

    order=df["cluster_insight"]
    .value_counts()
    .index
)

plt.title("Customer Segment Distribution")

plt.xlabel("Count")

plt.ylabel("Segment")

plt.show()


# ============================================================
# STEP 16 — REVIEW YEAR ANALYSIS
# ============================================================

plt.figure(figsize=(12,6))

sns.countplot(

    x="review_year",

    hue=cluster_column,

    data=df
)

plt.title("Review Trends by Year & Cluster")

plt.xticks(rotation=45)

plt.show()


# ============================================================
# STEP 17 — TOP BOOK TITLES
# ============================================================

top_books = (

    df["title"]
    .value_counts()
    .head(10)
)

print("\nTop Reviewed Books:")

print(top_books)


# ============================================================
# STEP 18 — TOP BOOK VISUALIZATION
# ============================================================

plt.figure(figsize=(12,6))

sns.barplot(

    x=top_books.values,

    y=top_books.index
)

plt.title("Top Reviewed Books")

plt.xlabel("Number of Reviews")

plt.ylabel("Book Title")

plt.show()


# ============================================================
# STEP 19 — CUSTOMER BEHAVIOR INSIGHTS
# ============================================================

print("\n======================================")
print("CUSTOMER BEHAVIOR INSIGHTS")
print("======================================")

print("\n1. Some clusters contain long analytical reviews")

print("\n2. Some customers prefer short emotional reactions")

print("\n3. Certain clusters indicate highly engaged readers")

print("\n4. NLP clustering successfully grouped similar review styles")

print("\n5. Word usage differs significantly across clusters")


# ============================================================
# STEP 20 — BUSINESS RECOMMENDATIONS
# ============================================================

print("\n======================================")
print("BUSINESS RECOMMENDATIONS")
print("======================================")

print("\n1. Target detailed reviewers for premium books")

print("\n2. Use emotional reviewers for marketing campaigns")

print("\n3. Detect repetitive review patterns for spam analysis")

print("\n4. Recommend books based on reviewer cluster")

print("\n5. Improve recommendation systems using customer segments")


# ============================================================
# STEP 21 — SAVE INSIGHT DATASET
# ============================================================

OUTPUT_PATH = (
    "../reports/"
    "business_insights_dataset.csv"
)

df.to_csv(
    OUTPUT_PATH,
    index=False
)

print("\nBusiness Insight Dataset Saved Successfully")


# ============================================================
# STEP 22 — SAVE EXECUTIVE SUMMARY
# ============================================================

summary_text = """

AMAZON BOOK REVIEWS CLUSTERING PROJECT
=====================================

PROJECT OBJECTIVE:
Cluster customer reviews using NLP and unsupervised learning.

KEY ACHIEVEMENTS:
- Implemented TF-IDF vectorization
- Applied PCA dimensionality reduction
- Compared KMeans, Agglomerative and DBSCAN
- Generated business-level customer segments
- Built customer review clusters

BUSINESS VALUE:
- Customer segmentation
- Review behavior understanding
- Recommendation improvement
- Review pattern discovery

BEST MODEL:
KMeans Clustering

"""

with open(

    "../reports/executive_summary.txt",

    "w",

    encoding="utf-8"

) as file:

    file.write(summary_text)

print("\nExecutive Summary Saved Successfully")


# ============================================================
# STEP 23 — FINAL CHECKPOINT SUMMARY
# ============================================================

print("\n================================================")
print("NOTEBOOK 06 COMPLETED SUCCESSFULLY")
print("================================================")

print("\nTasks Completed:")

print("1. Cluster Distribution Analysis")
print("2. Top Keyword Extraction")
print("3. Word Cloud Generation")
print("4. Review Pattern Analysis")
print("5. Reviewer Behavior Analysis")
print("6. Sample Review Exploration")
print("7. Customer Segment Identification")
print("8. Business Recommendations")
print("9. Executive Summary")
print("10. Insight Dataset Saved")

print("\nReady for Streamlit Dashboard Development")
```
